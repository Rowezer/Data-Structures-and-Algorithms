```python
#Бинарный поиск
import time
import random
def binary_search(array, num):
    array.sort()
    mid = 0
    start = 0
    end = len(array)
    step = 0

    while (start <= end):
        step = step+1
        mid = (start + end) // 2

        if num == array[mid]:
            return mid

        if num < array[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return print("Number not found")
def binary_search_add(array, num):
    array.sort()
    mid = 0
    start = 0
    end = len(array)
    step = 0

    while (start <= end):
        step = step+1
        mid = (start + end) // 2

        if num == array[mid]:
            return mid

        if num < array[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return array.append(num)
def binary_search_delete(array, num):
    array.sort()
    mid = 0
    start = 0
    end = len(array)
    step = 0

    while (start <= end):
        step = step+1
        mid = (start + end) // 2

        if num == array[mid]:
            return array.remove(num)

        if num < array[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return print("Number not found")
arr = [1,3,6,10,2,15,4]
arr.sort()
start_time = (time.perf_counter())
res=binary_search(arr,10)
end_time = (time.perf_counter())
print("Исходное значение:", 10)
print("Результат бинарного поиска: ", res)
print("Время бинарного поиска: {:.6f}ms".format((end_time-start_time)*100))
print("===========================")
start_time = (time.perf_counter())
result=arr.index(10)
end_time = (time.perf_counter())
print("Результат стандартного поиска: ", result)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Бинарное дерево
class Tree:
    right = None
    left = None
    
    def __init__(self, root:int, index:int, parent = None):
        self.root = root
        self.index = index
        self.parent = parent
        self.left = None
        self.right = None
        
    def add(self, value: int, index:int):     
        if value < self.root:
            if self.left == None:
                self.left = Tree(value, index, self)
            else:
                self.left.add(value, index)
        else:
            if self.right == None:
                self.right = Tree(value, index, self)
            else:
                self.right.add(value, index)
    
    def printTree(self):
        if self.left != None:
            self.left.printTree()
        if self.parent != None:
            print(self.root, "=>", self.parent.root)
        else:
            print(self.root, "<= корень")
        if self.right != None:
            self.right.printTree()
    
    def find(self, value:int):
        if self.root == value:
            return self.index
        elif value < self.root:
            if self.left == None:
                return -1
            else:
                return self.left.find(value)
        else:
            if self.right == None:
                return -1
            else:
                return self.right.find(value)
    
    def delete(self, value):
        if self.root == value:
            if self.right == None and self.left == None:
                newNode = None
            elif self.right == None:
                newNode = self.left
            elif self.left == None:
                newNode = self.right
            else:
                newNode = self.right.findMinEl()
                
            if self.parent != None:     
                if self.parent.right == self:
                    if newNode != None:
                        self.parent.right.root = newNode.root
                    else:
                        self.parent.right = None
                else:
                    if newNode != None:
                        self.parent.left.root = newNode.root
                    else:
                        self.parent.left = None
            elif newNode != None:
                self.root = newNode.root
            
            if newNode != None:
                newNode.delete(newNode.root)
            
        elif value < self.root:
            if self.left == None:
                return -1
            else:
                self.left.delete(value)
        else:
            if self.right == None:
                return -1
            else:
                self.right.delete(value)
                
    def findMinEl(self):
        if self.left == None:
            return self
        else:
            return self.left.findMinEl()
    
    def __cmp__(self, other):
        if self.root == other.root:
            return 0
        elif self.root < other.root:
            return -1
        else:
            return 1




mytree = Tree(arr[0], 0)
for i in range(1,len(arr)):
    mytree.add(arr[i],i)
start_time = (time.perf_counter())
res=mytree.find(10)
end_time = (time.perf_counter())
print("Исходное значение:", 10)
print("Результат бинарного поиска: ", res)
print("Время бинарного поиска: {:.6f}ms".format((end_time-start_time)*100))
print("===========================")
start_time = (time.perf_counter())
result=arr.index(10)
end_time = (time.perf_counter())
print("Результат стандартного поиска: ", result)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Интерполяционный
def InterpolationSearch(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    return "Not found"
def InterpolationSearch_delete(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            lys.remove(val)
            return "Число было удалено"
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    return "Not found"
def InterpolationSearch_append(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return "Число есть в списке"
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    lys.append(val)
    return "Число было добавлено в конец списка"
start_time = (time.perf_counter())
res=InterpolationSearch(arr,1)
end_time = (time.perf_counter())
print("Исходное значение:", 10)
print("Результат интерполяционного поиска: ", res)
print("Время интерполяционного поиска: {:.6f}ms".format((end_time-start_time)*100))
print("===========================")
start_time = (time.perf_counter())
result=arr.index(10)
end_time = (time.perf_counter())
print("Результат стандартного поиска: ", 10)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Простое рехэширование
class HashClass:
    keys = []
    hashes = []
    
    def __init__(self, arr : list):
        self.appendArray(arr)
        
    def appendKey(self, item, length = None):
        if (length == None):
            length = len(self.keys)
            
        self._checkKeyExists(item)
        temp = hash(item) % length
        
        if self._checkCollisions(temp):
            temp = self._simpleRehash(item, length)
            
        self.hashes.append(temp)
        self.keys.append(item)
        
    def printHashes(self):
        for i in range(len(self.keys)):
            print(self.hashes[i], self.keys[i])
        print()
        
    def appendArray(self, arr:list):
        for item in arr:
            self.appendKey(item, len(arr))
            
    def _simpleRehash(self, key, length):
        i = 1
        while True:
            temp = hash(key) % length + i
            if not self._checkCollisions(temp):
                return temp
            i = i + 1
    
    def _checkCollisions(self, hashValue):
        return hashValue in self.hashes

    def _checkKeyExists(self, key):
        if key in self.keys:
            return print("Key '{}' already exists".format(key))
values = ['pistolet', 'koridor', 'gornizon', 'doska', 'braslet']
hashes = HashClass(arr=values) 

hashes.printHashes()
hashes.appendKey('karandash')
hashes.appendKey('optika')
hashes.printHashes()
hashes.appendKey('optika')
```

```python
#Рехэширование с помощью псевдослучайных чисел
import random

class HashClass:
    keys = []
    hashes = []
    
    def __init__(self, arr : list):
        self.appendArray(arr)
        
    def appendKey(self, item, length = None):
        if (length == None):
            length = len(self.keys)
            
        self._checkKeyExists(item)
        temp = hash(item) % length
        
        if self._checkCollisions(temp):
            temp = self._pseudoRandRehash(item, length)
            
        self.hashes.append(temp)
        self.keys.append(item)
        
    def printHashes(self):
        for i in range(len(self.keys)):
            print(self.hashes[i], self.keys[i])
        print()
        
    def appendArray(self, arr:list):
        for item in arr:
            self.appendKey(item, len(arr))
            
    def _pseudoRandRehash(self, key, length : int):
        possibleValues = [i for i in range(length * 2)]
        random.shuffle(possibleValues)
        while True:
            i = possibleValues.pop()
            temp = hash(key) % length + i
            if not self._checkCollisions(temp):
                return temp
            
    def _checkCollisions(self, hashValue):
        return hashValue in self.hashes

    def _checkKeyExists(self, key):
        if key in self.keys:
              return print("Key '{}' already exists".format(key))
            
values = ['pistolet', 'koridor', 'gornizon', 'doska', 'braslet']
hashes = HashClass(arr=values)

hashes.printHashes()
hashes.appendKey('karandash')
hashes.appendKey('optika')
hashes.appendKey('dorama')
hashes.appendKey('kontakt')
hashes.printHashes()
hashes.appendKey('optika')

```

```python
#Метод цепочек
class ChainsMethod:
    _hashes = {}
    _values = {}
    _pointer = 0
    _N = 10
    
    def __init__(self, arr:dict = None):
        if arr != None:
            self.appendArray(arr)
    
    def append(self, key, value):
        keyHash = hash(key) % self._N
        if self._checkCollisions(keyHash):
            prevInd = self._hashes[keyHash]
            self._values[self._pointer] = [key, value, prevInd]
        else:
            self._values[self._pointer] = [key, value]
            
        self._hashes[keyHash] = self._pointer
        self._pointer += 1
            
    def appendArray(self, arr:dict):
        for key, value in arr.items():
            self.append(key, value)
            
    def find(self, key):
        keyHash = hash(key) % self._N
        if not keyHash in self._hashes:
            raise Exception('Item not found')
        index = self._hashes[keyHash]
        return self._recursiveFind(key, index)

    def _recursiveFind(self, key, index):
        if self._values[index][0] == key:
            return self._values[index][1]
        
        if len(self._values[index]) < 3:
            return print("Key '{}' not found".format(key))
            
        return self._recursiveFind(key, self._values[index][2])
            
    
    def printHashes(self):
        for key, value in self._values.items():
            print("{} - {}".format(key,value))
        
    def _checkCollisions(self, keyHash):
        return keyHash in self._hashes
arr = {"key1" : "value1", "key2" : "value2", "key3" : "value3"}
chains = ChainsMethod(arr)
chains.printHashes()
print('New HashTable')
chains.append("newKey", "newValue")
chains.printHashes()
print('find "key2"')
print("result: ",chains.find("key2"))
print('find "key"')
print("result: ",chains.find("key"))
```

Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
global queen
global number
EIGHT = 8
queen = [None] * 8
number = 0

def print_table():
    global number
    x = y = 0
    number += 1
    for x in range(EIGHT):
        for y in range(EIGHT):
            if x == queen[y]:
                print('\033[91m' + ' [Q] ', end='')
            else:
                print('\033[94m' + ' [0] ', end='')
        print('\n')
    print('\n\n')
    
def attack(row, col):
    global queen
    i = 0
    atk = 0
    offset_row = offset_col = 0
    while (atk != 1) and i < col:
        offset_col = abs(i - col)
        offset_row = abs(queen[i] - row)
        if queen[i] == row or offset_row == offset_col:
            atk = 1
        i = i +1
    return atk

def decide_position(value):
    global queen
    i = 0
    while i < EIGHT:
        if attack(i,value) != 1:
            queen[value] = i
            if value == 7:
                print_table()
            else:
                decide_position(value + 1)
        i = i + 1
        
decide_position(0)
```
