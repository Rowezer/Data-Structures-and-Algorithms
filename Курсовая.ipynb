–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ m, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ–±–µ—Ä –≥—Ä–∞—Ñ–∞.

–î–∞–ª–µ–µ –≤—ã–≤–µ–¥–∏—Ç–µ m —Å—Ç—Ä–æ–∫, –∫–∞–∂–¥–∞—è –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–ª–∞ a –∏ b (0‚Äâ‚â§‚Äâa‚Äâ‚â§‚Äân‚Äâ-‚Äâ1, 0‚Äâ‚â§‚Äâb‚Äâ‚â§‚Äân‚Äâ-‚Äâ1), —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ —Ä–µ–±—Ä—É (a,‚Äâb).

–†—ë–±—Ä–∞ –º–æ–≥—É—Ç –±—ã—Ç—å –≤—ã–≤–µ–¥–µ–Ω—ã –≤ –ª—é–±–æ–º –ø–æ—Ä—è–¥–∫–µ; –≤–µ—Ä—à–∏–Ω—ã –æ–¥–Ω–æ–≥–æ —Ä–µ–±—Ä–∞ —Ç–∞–∫–∂–µ –º–æ–≥—É—Ç –±—ã—Ç—å –≤—ã–≤–µ–¥–µ–Ω—ã –≤ –ª—é–±–æ–º –ø–æ—Ä—è–¥–∫–µ.

```python
n = int(input())
if n <1 or n>216:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    v = []
    deg1 = []
    degsum = 0
    has_v = [True]*n
    for i in range(n):
        d, s = map(int, input().split())
        degsum += d
        if d == 1: deg1.append(i)
        if d == 0: has_v[i] = False
        v.append([d, s])
 
    if d<0 or d>n-1 or s<0 or s>216:
        print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    else:
        edge = []
        while deg1 != []:
            f = deg1.pop()
            if has_v[f] is False: continue
            v[f][0] -= 1
            has_v[f] = False
            t = v[f][1]
            edge.append([f, t])
            v[t][0] -= 1
            v[t][1] ^= f
            if v[t][0] == 1:
                deg1.append(t)
            elif v[t][0] == 0:
            has_v[t] = False
        print("---result------")
        print(degsum//2)
        for f, t in edge:
            print(f, t)
```

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ n, m –∏ k (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâm‚Äâ‚â§‚Äâ109, 1‚Äâ‚â§‚Äâk‚Äâ‚â§‚Äân) ‚Äî —á–∏—Å–ª–æ —Ö–æ–±–±–∏—Ç–æ–≤, —á–∏—Å–ª–æ –ø–æ–¥—É—à–µ–∫ –∏ –Ω–æ–º–µ—Ä –∫—Ä–æ–≤–∞—Ç–∏ –§—Ä–æ–¥–æ.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –ø–æ–¥—É—à–µ–∫, –∫–æ—Ç–æ—Ä—ã–µ –§—Ä–æ–¥–æ –º–æ–∂–µ—Ç –ø–æ–ª–æ–∂–∏—Ç—å —Å–µ–±–µ, —á—Ç–æ–±—ã –Ω–∏–∫—Ç–æ –Ω–µ –æ–±–∏–¥–µ–ª—Å—è.

–ü—Ä–∏–º–µ—Ä—ã

4 6 2
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
2

```python
n, m, k = map(int, input().split())
if n<1 or n>m or m>109 or k<1 or k>n:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    m -= n
    d = 0
    k -= 1
    result = 1
    while m > 1 and d != max(k, n - k - 1):
        result += 1
        m -= 1
        left = min(d, k)
        right = min(d, n - k - 1)
        d += 1
        m -= left
        m -= right
    result += m // n
    print("---result------")
    print(result)
```

–ù–∞–π–¥–∏—Ç–µ —Ç–∞–∫–æ–π —Å–ø–æ—Å–æ–± –∏–∑–º–µ–Ω–∏—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, —á—Ç–æ —Å—É–º–º–∞—Ä–Ω–æ–µ –≤—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø—ã –±—É–¥–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ. –í–æ–∑–º–æ–∂–Ω–æ, —á—Ç–æ –∏–Ω–æ–≥–¥–∞ –ª—É—á—à–µ –æ—Å—Ç–∞–≤–∏—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É –Ω–µ—Ç—Ä–æ–Ω—É—Ç–æ–π. (–ï—Å–ª–∏ —É –ª–∞–º–ø—ã –≤–∫–ª—é—á–µ–Ω —Å–≤–µ—Ç —Å –º–æ–º–µ–Ω—Ç–∞ ùë• –ø–æ –º–æ–º–µ–Ω—Ç ùë¶, —Ç–æ–≥–¥–∞ –ª–∞–º–ø–∞ –±—É–¥–µ—Ç –≥–æ—Ä–µ—Ç—å –≤—Ä–µ–º—è, —Ä–∞–≤–Ω–æ–µ ùë¶‚àíùë•. –û—Ç—Ä–µ–∑–∫–∏ –≤–∫–ª—é—á–µ–Ω–Ω–æ–≥–æ —Å–≤–µ—Ç–∞ —Å—É–º–º–∏—Ä—É—é—Ç—Å—è).

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —á–∏—Å–ª–∞ ùëõ –∏ ùëÄ (1‚â§ùëõ‚â§105, 2‚â§ùëÄ‚â§109) ‚Äî –¥–ª–∏–Ω–∞ –ø—Ä–µ–¥—É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω–æ–π –ø—Ä–æ–≥—Ä–∞–º–º—ã ùëé –∏ –º–æ–º–µ–Ω—Ç –æ—Ç–∫–ª—é—á–µ–Ω–∏—è –ø–∏—Ç–∞–Ω–∏—è –ª–∞–º–ø—ã.

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç ùëõ —á–∏—Å–µ–ª —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª ùëé1,ùëé2,‚Ä¶,ùëéùëõ (0<ùëé1<ùëé2<‚ãØ<ùëéùëõ<ùëÄ) ‚Äî –ø—Ä–æ–≥—Ä–∞–º–º–∞ ùëé.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ –≤—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è, –µ—Å–ª–∏ –í—ã –º–æ–∂–µ—Ç–µ —Å–¥–µ–ª–∞—Ç—å –Ω–µ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è, –æ–ø–∏—Å–∞–Ω–Ω–æ–≥–æ –≤—ã—à–µ.

–ü—Ä–∏–º–µ—Ä—ã
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
3 10
4 6 7
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
8

```python
m,n=map(int,input().split())
a=[0]+[int(x) for x in input().split()]+[n]
s=k=t=0
abool=False
if n<1 or n>105 or m<2 or m>109:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    for i in range (len(a)-1):
        for j in range (i,len(a)):
            if a[i]>a[j]:
                    abool=True
                    break
    if abool is True:
        print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    else:
        for i in range(1,len(a),2):
            ti=a[i]-a[i-1]
            t+=ti
        s=t
        for i in range(2,len(a),2):
            k+=a[i-1]-a[i-2]
            if a[i]>a[i-1]+1:
                ki=t-k
                s=max(s,k+n-a[i-1]-1-ki)
        print("---result------")
        print(s)
```

–£ –í–∞—Å–∏ –µ—Å—Ç—å n –ø—Ä–µ–¥–º–µ—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –ª–µ–∂–∞—Ç –≤ —Ä—è–¥. –ü—Ä–µ–¥–º–µ—Ç—ã –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–Ω—É–º–µ—Ä–æ–≤–∞–Ω—ã —á–∏—Å–ª–∞–º–∏ –æ—Ç 1 –¥–æ n —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ —Å–∞–º—ã–π –ª–µ–≤—ã–π –ø—Ä–µ–¥–º–µ—Ç –∏–º–µ–µ—Ç –Ω–æ–º–µ—Ä 1, —Å–∞–º—ã–π –ø—Ä–∞–≤—ã–π ‚Äî –Ω–æ–º–µ—Ä n. –ö–∞–∂–¥—ã–π –ø—Ä–µ–¥–º–µ—Ç –∏–º–µ–µ—Ç —Å–≤–æ–π –≤–µ—Å, i-—Ç—ã–π –∏–∑ –Ω–∏—Ö –≤–µ—Å–∏—Ç wi –∫–∏–ª–æ–≥—Ä–∞–º–º.

–í–∞—Å–µ –Ω—É–∂–Ω–æ —Å–æ–±—Ä–∞—Ç—å –≤—Å–µ —ç—Ç–∏ –ø—Ä–µ–¥–º–µ—Ç—ã, –æ–¥–Ω–∞–∫–æ –æ–Ω –Ω–µ –±—É–¥–µ—Ç –¥–µ–ª–∞—Ç—å —ç—Ç–æ–≥–æ —Å–∞–º, –∞ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç —Å–≤–æ–µ–≥–æ –Ω–æ–≤–æ–≥–æ —Ä–æ–±–æ—Ç–∞. –£ —Ä–æ–±–æ—Ç–∞ –µ—Å—Ç—å –¥–≤–µ —Ä–∞–∑–Ω—ã–µ —Ä—É–∫–∏ ‚Äî –ª–µ–≤–∞—è –∏ –ø—Ä–∞–≤–∞—è. –†–æ–±–æ—Ç –º–æ–∂–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –≤—ã–ø–æ–ª–Ω—è—Ç—å –¥–µ–π—Å—Ç–≤–∏—è:

–í–∑—è—Ç—å –ª–µ–≤–æ–π —Ä—É–∫–æ–π —Å–∞–º—ã–π –ª–µ–≤—ã–π –∏–∑ –ø—Ä–µ–¥–º–µ—Ç–æ–≤, –∑–∞—Ç—Ä–∞—Ç–∏–≤ –Ω–∞ —ç—Ç–æ wi‚Äâ¬∑‚Äâl –µ–¥–∏–Ω–∏—Ü —ç–Ω–µ—Ä–≥–∏–∏ (–≥–¥–µ wi ‚Äî –≤–µ—Å —Å–∞–º–æ–≥–æ –ª–µ–≤–æ–≥–æ –ø—Ä–µ–¥–º–µ—Ç–∞, l ‚Äî –∑–∞–¥–∞–Ω–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä), –µ—Å–ª–∏ –ø—Ä–µ–¥—ã–¥—É—â–µ–µ –¥–µ–π—Å—Ç–≤–∏–µ –±—ã–ª–æ —Ç–∞–∫–∏–º –∂–µ (–ª–µ–≤–æ–π —Ä—É–∫–æ–π), —Ç–æ —Ä–æ–±–æ—Ç —Ç—Ä–∞—Ç–∏—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ Ql –µ–¥–∏–Ω–∏—Ü —ç–Ω–µ—Ä–≥–∏–∏. –í–∑—è—Ç—å –ø—Ä–∞–≤–æ–π —Ä—É–∫–æ–π —Å–∞–º—ã–π –ø—Ä–∞–≤—ã–π –∏–∑ –ø—Ä–µ–¥–º–µ—Ç–æ–≤, –∑–∞—Ç—Ä–∞—Ç–∏–≤ –Ω–∞ —ç—Ç–æ wj‚Äâ¬∑‚Äâr –µ–¥–∏–Ω–∏—Ü —ç–Ω–µ—Ä–≥–∏–∏ (–≥–¥–µ wj ‚Äî –≤–µ—Å —Å–∞–º–æ–≥–æ –ø—Ä–∞–≤–æ–≥–æ –ø—Ä–µ–¥–º–µ—Ç–∞, r ‚Äî –∑–∞–¥–∞–Ω–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä), –µ—Å–ª–∏ –ø—Ä–µ–¥—ã–¥—É—â–µ–µ –¥–µ–π—Å—Ç–≤–∏–µ –±—ã–ª–æ —Ç–∞–∫–∏–º –∂–µ (–ø—Ä–∞–≤–æ–π —Ä—É–∫–æ–π), —Ç–æ —Ä–æ–±–æ—Ç —Ç—Ä–∞—Ç–∏—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ Qr –µ–¥–∏–Ω–∏—Ü —ç–Ω–µ—Ä–≥–∏–∏. –†–∞–∑—É–º–µ–µ—Ç—Å—è, –í–∞—Å—è —Ö–æ—á–µ—Ç –∑–∞–ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞—Ç—å —Ä–æ–±–æ—Ç–∞ —Ç–∞–∫, —á—Ç–æ–±—ã —Ç–æ—Ç –ø–æ—Ç—Ä–∞—Ç–∏–ª –∫–∞–∫ –º–æ–∂–Ω–æ –º–µ–Ω—å—à–µ —ç–Ω–µ—Ä–≥–∏–∏. –° —ç—Ç–æ–π –∑–∞–¥–∞—á–µ–π –æ–Ω –∏ –æ–±—Ä–∞—Ç–∏–ª—Å—è –∑–∞ –ø–æ–º–æ—â—å—é –∫ –≤–∞–º. –í–∞—à–∞ –∑–∞–¥–∞—á–∞ ‚Äî –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–Ω–µ—Ä–≥–∏–∏, –∫–æ—Ç–æ—Ä–æ–µ –ø–æ—Ç—Ä–∞—Ç–∏—Ç —Ä–æ–±–æ—Ç, —á—Ç–æ–±—ã —Å–æ–±—Ä–∞—Ç—å –≤—Å–µ –ø—Ä–µ–¥–º–µ—Ç—ã.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –ø—è—Ç—å —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª n,‚Äâl,‚Äâr,‚ÄâQl,‚ÄâQr (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ105;‚Äâ1‚Äâ‚â§‚Äâl,‚Äâr‚Äâ‚â§‚Äâ100;‚Äâ1‚Äâ‚â§‚ÄâQl,‚ÄâQr‚Äâ‚â§‚Äâ104).

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è n —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª w1,‚Äâw2,‚Äâ...,‚Äâwn (1‚Äâ‚â§‚Äâwi‚Äâ‚â§‚Äâ100).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

–ü—Ä–∏–º–µ—Ä—ã
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
3 4 4 19 1
42 3 99
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
576

```python
n,l,r,ql,qr = map(int,input().split())
wbool=False
if n<1 or n>105 or l<1 or r>100 or ql<1 or qr>104:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    w = list(map(int, input().split()))
    for i in range (len(w)):
        if w[i]<1 or w[i]>100:
            print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
            wbool=True
            break
    if wbool is False:
        s = [0, *w]
        for i in range(1, n+1): s[i] += s[i-1]

        c_min =  s[n]*r + (n-1)*qr
        for i in range(1, n+1):
            cc = l*s[i] + (s[n]- s[i])*r
            if 2*i < n:
                cc += (n-2*i-1)*qr
            elif 2*i > n:
                cc += (i-(n-i) -1)*ql
            c_min = min(c_min, cc)
        print("---result------")
        print(c_min)
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í—ã–≤–µ–¥–∏—Ç–µ n‚Äâ+‚Äâ1 —á–∏—Å–µ–ª a0,‚Äâa1,‚Äâ...,‚Äâan, –≥–¥–µ a0 ‚Äî —Å–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –≤ –Ω–∞—á–∞–ª–µ, a1 ‚Äî —Å–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è –ø–æ—Å–ª–µ –∑–∞–º–µ–Ω—ã –æ–¥–Ω–æ–π –º–æ–Ω–µ—Ç—ã –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ.

```python
n = int(input())
if n<1 or n>300000:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    plist = list(map(int, input().split(' ')))
    for i in range(len(plist)):
        if p[i]<1 or p[i]>n:
            print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
            pbool=False
            break
    if pbool is True:
        pos = n
        a = [0 for i in range(n+1)]
        res = 1
        ans = [1]
        for x in plist:
            a[x] = 1
            res += 1
            while a[pos]==1:
                pos -= 1
                res -= 1
            ans.append(res)
        print("---result------")
        print (' '.join(map(str, ans)))
```

–ü–æ–º–æ–≥–∏—Ç–µ –µ–º—É, –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –æ–ø–∏—Å–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª x –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π –ø–∞—Ä—ã l –∏ r.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–∞ –ø–∞—Ä–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª l –∏ r (1‚Äâ‚â§‚Äâl‚Äâ‚â§‚Äâr‚Äâ‚â§‚Äâ1018) ‚Äî –≥—Ä–∞–Ω–∏—Ü—ã –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞.

–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–ø–µ—Ü–∏—Ñ–∏–∫–∞—Ç–æ—Ä %lld –¥–ª—è —á—Ç–µ–Ω–∏—è –∏–ª–∏ –∑–∞–ø–∏—Å–∏ 64-—Ö –±–∏—Ç–æ–≤—ã—Ö —á–∏—Å–µ–ª –Ω–∞ –°++. –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–æ—Ç–æ–∫–∏ cin, cout –∏–ª–∏ —Å–ø–µ—Ü–∏—Ñ–∏–∫–∞—Ç–æ—Ä %I64d.

–ü—Ä–∏–º–µ—Ä—ã
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
2 47
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

```python
def find(x):
    if(x<10):
        return x 
    ans=x//10-1 
    ans+=9
    if(str(x)[0]<=str(x)[-1]):
        ans+=1 
    return ans
l,r=map(int,input().split())
if l<1 or l>r or r>1018:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    print("---result------")
    print( find(r) - find(l-1))
```

## –ó–∞–¥–∞—á–∞ 8

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∑–∞–¥–∞–Ω–æ —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (3‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –º–∞—Å—Å–∏–≤–µ a. –í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ n –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ai (1‚Äâ‚â§‚Äâai‚Äâ‚â§‚Äâ109) ‚Äî —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö —Ç—Ä–æ–µ–∫ (i,‚Äâ j,‚Äâ k), —á—Ç–æ i,‚Äâ j –∏ k ‚Äî –ø–æ–ø–∞—Ä–Ω–æ —Ä–∞–∑–ª–∏—á–Ω—ã, –∞ ai¬∑aj¬∑ak ‚Äî –º–∏–Ω–∏–º–∞–ª—å–Ω–æ.

–ü—Ä–∏–º–µ—Ä—ã
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
4
1 1 1 1
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
4

```python
a=int(input())
bbool=True
if a<3 or a>105:
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
else:
    b = list(map(int, input().split()))
    s=[]
    if a!=len(b):
        print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    else:
        for i in range(len(b)):
            if b[i]<1 or b[i]>109:
                print("–í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
                break
                bbool=False
            s.append(b[i])
        if bbool is True: 
            s.sort()
            f1=b.count(s[0])
            b.remove(s[0])
            f2=b.count(s[1])
            b.remove(s[1])
            f3=b.count(s[2])
            b.remove(s[2])
            f4=f1*f2*f3
            if s[1]==s[2] and s[2]==s[0]:
                f4=f4//6
            if s[1]==s[2] and s[2]!=s[0]:
                f4=f4//2
            if s[1]==s[0] and s[0]!=s[2]:
                f4=f4//2
            if s[2]==s[0] and s[1]!=s[2]:
                f4=f4//2
            print("---result------")
            print(f4)
```
